```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
df = pd.read_csv('../data/train_data.csv')
print(f'df.shape: {df.shape}')
display(df.head())
```

```python
def plot_categorical_distribution(df, col):
    value_counts = df[col].value_counts()
    plt.figure(figsize=(10, 6))
    sns.barplot(x=value_counts.values, y=value_counts.index, palette="viridis")
    plt.xlabel("Точка отказа")
    plt.ylabel("Frequency")
    plt.title(f"Frequency of Unique Values in {col}")
    plt.xticks(rotation=90)
    plt.show()

plot_categorical_distribution(df, 'Точка отказа')
```

```python
plot_categorical_distribution(df, 'Тип оборудования')
```

```python
import pandas as pd
df = pd.read_csv('../data/data_date.csv')
df['created_at'] = pd.to_datetime(df['created_at'], errors='coerce').dt.date
```

```python
df['created_at'][0]
```

```python
import pandas as pd
import plotly.express as px

# Ensure 'created_at' is in datetime format
df['created_at'] = pd.to_datetime(df['created_at'], errors='coerce')

# Drop rows where 'created_at' is invalid (NaT)
df = df.dropna(subset=['created_at'])

# Group by 'created_at' date and count occurrences
day_count = df.groupby(df['created_at'].dt.date).size().reset_index(name='Количество')
day_count = day_count.sort_values('created_at')

# Plot the time series for the number of appeals
fig1 = px.line(day_count, x='created_at', y='Количество', title='Кол-во обращений в течение времени')
fig1.update_xaxes(type='category')  # Set x-axis to categorical for days
fig1.update_layout(xaxis_title='Дата', yaxis_title='Кол-во')

# Now, group by both 'created_at' date and 'Точка отказа' to get the counts per category
day_point_count = df.groupby([df['created_at'].dt.date, 'Тип оборудования']).size().reset_index(name='Количество')

# Plot the time series with breakdown by 'Точка отказа'
fig2 = px.line(day_point_count, x='created_at', y='Количество', color='Тип оборудования', title='Кол-во обращений по точке отказа')
fig2.update_xaxes(type='category')  # Set x-axis to categorical for days
fig2.update_layout(xaxis_title='Дата', yaxis_title='Кол-во', legend_title='Тип оборудования')

# Show both figures
fig2
```
